# Archivo Excel

### Importamos los datasets y generamos un DataFrame para cada uno, verificando posteriormente la cantidad de filas y columnas correspondientes a cada conjunto de datos.

```python
url_clinical_amps = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/6e21d108d3f7d3319102a1a0d7e4ba585136e01d/Base%20de%20datos/DRAMP/clinical_amps.xlsx"

df_clinical_amps = pd.read_excel(url_clinical_amps)
df_clinical_amps.to_csv("clinical_amps.csv", index=False)

df_clinical_amps.head(2)
```

```python
df_clinical_amps.shape
```

```python
url_expanded_amps = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/cdeebc4d99639d5cd5c37c6dee38f4e1355e772c/Base%20de%20datos/DRAMP/expanded_amps.xlsx"

df_expanded_amps = pd.read_excel(url_expanded_amps)
df_expanded_amps.to_csv("expanded_amps.csv", index=False)

df_expanded_amps.head(2)
```

```python
df_expanded_amps.shape
```

```python
url_general_amps = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/cdeebc4d99639d5cd5c37c6dee38f4e1355e772c/Base%20de%20datos/DRAMP/general_amps.xlsx"

df_general_amps = pd.read_excel(url_general_amps)
df_general_amps.to_csv("general_amps.csv", index=False)

df_general_amps.head(2)
```

```python
df_general_amps.shape
```

```python
url_patent_amps = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/cdeebc4d99639d5cd5c37c6dee38f4e1355e772c/Base%20de%20datos/DRAMP/patent_amps.xlsx"

df_patent_amps = pd.read_excel(url_patent_amps)
df_patent_amps.to_csv("patent_amps.csv", index=False)

df_patent_amps.head(2)
```

```python
df_patent_amps.shape
```

```python
url_specific_amps = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/cdeebc4d99639d5cd5c37c6dee38f4e1355e772c/Base%20de%20datos/DRAMP/specific_amps.xlsx"
df_specific_amps = pd.read_excel(url_specific_amps)
df_specific_amps.to_csv("specific_amps.csv", index=False)

df_specific_amps.head(2)
```

```python
df_specific_amps.shape
```

```python
url_stability_amps = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/3365142f821c37a43b7a4058e8eafa20497a7593/Base%20de%20datos/DRAMP/stability_amps.xlsx"

df_stability_amps = pd.read_excel(url_stability_amps)
df_stability_amps.to_csv("stability_amps.csv", index=False)

df_stability_amps.head(2)
```

```python
df_stability_amps.shape
```

### Concatenamos los DataFrames

```python
df_concat = pd.concat(
    [df_clinical_amps, df_expanded_amps,df_general_amps,df_patent_amps, df_specific_amps,df_stability_amps], # Lista de DataFrames a concatenar
    axis=0, # Eje: 0 para filas (vertical, por defecto), 1 para columnas (horizontal)
    ignore_index=False # True si quieres un nuevo índice consecutivo
)
df_concat.head(2)
```

### Escogemos solo las columnas de interes (columnas con secuencia y actividad)

```python
df_DRAMP_activity = df_concat[["Sequence", "Activity"]]
df_DRAMP_activity.head()
```

### Filtramos la columna Activity para que solo indique los que contienen la actividad antifungal

```python
df_DRAMP_activity_antifungal = df_DRAMP_activity[df_DRAMP_activity['Activity'].str.contains("Antifungal|Antifunga|Antifugal", case=False, na=False)].copy()
df_DRAMP_activity_antifungal.head()
```

```python
df_DRAMP_activity_antifungal.shape
```

### Creamos una columna llamada label y accinamos 1, considerando que todas estas secuencias presentan la actividad antifungal

```python
df_DRAMP_activity_antifungal['label'] = 1
df_DRAMP_activity_antifungal
```

### Finalmente solo nos quedamos con las columnas sencuencia y label

```python
df_DRAMP_antifungal = df_DRAMP_activity_antifungal[["Sequence", "label"]]
df_DRAMP_antifungal.head()
```

### Descargamos CSV final

```python
dataframe_a_descargar = df_DRAMP_antifungal
dataframe_a_descargar.to_csv(
    'DRAMP_antifungal.csv',  # Nombre que tendrá el archivo
    index=False                        # Importante: No incluir el índice de filas de Pandas
)
files.download('DRAMP_antifungal.csv')
```
