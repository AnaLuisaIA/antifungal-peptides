### Importamos librer√≠a pandas

```python
import pandas as pd
```

### Archivos CSV

### Importamos Datasets, convertimos a Dataframe y revisamos la cantidad de filas y columnas que posee cada uno.

```python
url_DeepAFP_main_train = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/011419fe410243c5bcb20c94c53883737c25a130/Base%20de%20datos/DeepAFP/DeepAFP-main-train.csv"

df_DeepAFP_main_train = pd.read_csv(url_DeepAFP_main_train)
df_DeepAFP_main_train.to_csv("DeepAFP-main-train.csv", index=False)

df_DeepAFP_main_train.head()
```

```python
url_DeepAFP_main_test = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/011419fe410243c5bcb20c94c53883737c25a130/Base%20de%20datos/DeepAFP/DeepAFP-main-test.csv"

df_DeepAFP_main_test = pd.read_csv(url_DeepAFP_main_test)
df_DeepAFP_main_test.to_csv("DeepAFP-main-test.csv", index=False)

df_DeepAFP_main_test.head()
```

```python
url_DeepAFP_Set1_traint = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/011419fe410243c5bcb20c94c53883737c25a130/Base%20de%20datos/DeepAFP/DeepAFP-Set1-train.csv"

df_DeepAFP_Set1_traint = pd.read_csv(url_DeepAFP_Set1_traint)
df_DeepAFP_Set1_traint.to_csv("DeepAFP-Set1-train.csv", index=False)

df_DeepAFP_Set1_traint.head()
```

```python
url_DeepAFP_Set1_test = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/011419fe410243c5bcb20c94c53883737c25a130/Base%20de%20datos/DeepAFP/DeepAFP-Set1-test.csv"

df_DeepAFP_Set1_test = pd.read_csv(url_DeepAFP_Set1_test)
df_DeepAFP_Set1_test.to_csv("DeepAFP-Set1-test.csv", index=False)

df_DeepAFP_Set1_test.head()
```

```python
url_DeepAFP_Set2_traint  = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/011419fe410243c5bcb20c94c53883737c25a130/Base%20de%20datos/DeepAFP/DeepAFP-Set2-train.csv"

df_DeepAFP_Set2_traint = pd.read_csv(url_DeepAFP_Set2_traint)
df_DeepAFP_Set2_traint.to_csv("DeepAFP-Set2-train.csv", index=False)

df_DeepAFP_Set2_traint.head()
```

```python
url_DeepAFP_Set2_test = "https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/011419fe410243c5bcb20c94c53883737c25a130/Base%20de%20datos/DeepAFP/DeepAFP-Set2-test.csv"

df_DeepAFP_Set2_test = pd.read_csv(url_DeepAFP_Set2_test)
df_DeepAFP_Set2_test.to_csv("DeepAFP-Set2-test.csv", index=False)

df_DeepAFP_Set2_test.head()
```

### Concatenamos DataFrames

```python
df_concat = pd.concat(
    [df_DeepAFP_main_train, df_DeepAFP_main_test,df_DeepAFP_Set1_traint,df_DeepAFP_Set1_test, df_DeepAFP_Set2_traint,df_DeepAFP_Set2_test],
    axis=0,
    ignore_index=False
)
df_concat.head()
```

### Renombramos la columna sequence, para que siga el mismo formato de los otros Dataframes

```python
df_concat = df_concat.rename(columns={'sequence': 'Sequence'})
df_concat.head()
```

### Escogemos solo columnas Sequence y label

```python
df_DeepAFP = df_concat[["Sequence", "label"]]
df_DeepAFP.head()
```
