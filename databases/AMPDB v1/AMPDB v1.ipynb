```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/refs/heads/main/databases/AMPDB%20v1/Anti-candida%20dataset.tsv'
df_candida = pd.read_csv(url, sep='\t')
```

```python
df_candida.columns = df_candida.columns.str.strip()
```

```python
df_candida.head()
```

```python
url = 'https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/refs/heads/main/databases/AMPDB%20v1/Anti-yeast%20dataset.tsv'
df_yeast = pd.read_csv(url, sep='\t')
```

```python
df_yeast.columns = df_yeast.columns.str.strip()
```

```python
df_yeast.head()
```

```python
url = 'https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/refs/heads/main/databases/AMPDB%20v1/Fungicide%20dataset.tsv'
df_fungicide = pd.read_csv(url, sep='\t')
```

```python
df_fungicide.columns = df_fungicide.columns.str.strip()
```

```python
df_fungicide.columns
```

```python
# Seleccionar la secuencia
candida_sequences = df_candida[['Sequence']]
yeast_sequences = df_yeast[['Sequence']]
fungicide_sequences = df_fungicide[['Sequence']]
```

```python
# Agregar una columna de label (si tiene actividad antifúngica o no)
candida_sequences['label'] = 1
yeast_sequences['label'] = 1
fungicide_sequences['label'] = 1
```

```python
# Concatenar los dataframes
df_combined = pd.concat([candida_sequences, yeast_sequences, fungicide_sequences])
```

```python
# Display the head of the new dataframe
df_combined.shape
```

## Verificar duplicados

```python
df_combined["Sequence"].unique().shape
```

```python
df = df_combined.drop_duplicates().copy()
```

```python
df.shape
```

## Verificar canónicos

```python
# Definir aminoácidos canónicos
aminoacidos = set("ACDEFGHIKLMNPQRSTVWY")
```

```python
# Verificar cuáles secuencias válidas
df["is_valid"] = [set(seq).issubset(aminoacidos) for seq in df["Sequence"]]
```

```python
no_validas = df[~df["is_valid"]]
print("Secuencias inválidas:", len(no_validas))
```

```python
df_validas = df[df["is_valid"]].drop(columns="is_valid")
print("Secuencias válidas restantes:", len(df_validas))
```

```python
display(df_validas)
```

```python
df_validas = df_validas.rename(columns={'Sequence': 'sequence'})
display(df_validas.head())
```

```python
df_validas.to_csv('AMPDBv1_labeled.csv', index=False, header=True)
```

## Obtener metadatos

```python
import json
```

```python
def export_json(path_to_export, data_to_export):
    with open(path_to_export, 'w') as doc_export:
        json.dump(
            data_to_export,
            doc_export,
            indent=4,
            default=str,
            ensure_ascii=False)
```

```python
def create_metada_with_multiple_values(df_metada_filter, full_df):
    dict_metadata = {}

    for column in df_metada_filter.columns:
        values = df_metada_filter[column].unique().tolist()

        if len(values)>1:
            values = [str(value) for value in values]
            values = ";".join(values)
            dict_metadata.update({column:values})
        else:
            dict_metadata.update({column:values[0]})

    dict_metadata.update({
        "number_of_sequences" : len(full_df)
    })

    return dict_metadata
```

```python
def read_metadata(path_data, name_source):
    df_metada = pd.read_excel(path_data)
    df_metada_filter = df_metada[df_metada["name source"] == name_source]
    df_metada_filter = df_metada_filter[['type source',
                                         'estatico-dinamico',
                                         'licencia',
                                         'año de publicación',
                                         'fecha ultima actualizacion',
                                         'download date',
                                         'formato',
                                         'peptide property',
                                         'informacion del dataset',
                                         'unidad de medida',
                                         'Construccion de dataset negativos',
                                         'repositorio o servidor',
                                         'Publicacion',]]
    return df_metada_filter
```

```python
df_metada_filter = read_metadata("/content/description_raw_data.xlsx", "AMPDB v1")
```

```python
dict_metadata = create_metada_with_multiple_values(df_metada_filter, df_validas)
```

```python
export_json("metadata.json", dict_metadata)
```
