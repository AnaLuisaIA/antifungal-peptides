# Información sobre el conjunto de datos

Está relacionado con la publicación:  
[Developing anti-microbial peptide database version 1 to provide comprehensive and exhaustive resource of manually curated AMPs](https://www.nature.com/articles/s41598-023-45016-3)

Incluye ejemplos positivos de:
- Antifungal
- Anticandida
- Antiyeast

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/refs/heads/main/AMPDB%20v1/Anti-candida%20dataset.tsv'
df_candida = pd.read_csv(url, sep='\t')
```

```python
df_candida.columns = df_candida.columns.str.strip()
```

```python
df_candida.head(5)
```

```python
url = 'https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/refs/heads/main/AMPDB%20v1/Anti-yeast%20dataset.tsv'
df_yeast = pd.read_csv(url, sep='\t')
```

```python
df_yeast.columns = df_yeast.columns.str.strip()
```

```python
df_yeast.head()
```

```python
url = 'https://raw.githubusercontent.com/AnaLuisaIA/antifungal-peptides/refs/heads/main/AMPDB%20v1/Fungicide%20dataset.tsv'
df_fungicide = pd.read_csv(url, sep='\t')
```

```python
df_fungicide.columns = df_fungicide.columns.str.strip()
```

```python
df_fungicide.columns
```

```python
# Seleccionar la secuencia
candida_sequences = df_candida[['Sequence']]
yeast_sequences = df_yeast[['Sequence']]
fungicide_sequences = df_fungicide[['Sequence']]
```

```python
# Agregar una columna de label (si tiene actividad antifúngica o no)
candida_sequences['label'] = 1
yeast_sequences['label'] = 1
fungicide_sequences['label'] = 1
```

```python
# Concatenar los dataframes
df_combined = pd.concat([candida_sequences, yeast_sequences, fungicide_sequences])
```

```python
# Display the head of the new dataframe
df_combined.describe()
```

```python
df_combined.to_csv('AMPDB_combinado.csv', index=False, header=True)
```
