### Concatenamos los DataFrames

```python
df_concat = pd.concat(
    [df_clinical_amps, df_expanded_amps,df_general_amps,df_patent_amps, df_specific_amps,df_stability_amps], # Lista de DataFrames a concatenar
    axis=0, # Eje: 0 para filas (vertical, por defecto), 1 para columnas (horizontal)
    ignore_index=False # True si quieres un nuevo índice consecutivo
)
df_concat.head(2)
```

### Escogemos solo las columnas de interes (columnas con secuencia y actividad)

```python
df_DRAMP_activity = df_concat[["Sequence", "Activity"]]
df_DRAMP_activity.head()
```

### Filtramos la columna Activity para que solo indique los que contienen la actividad antifungal

```python
df_DRAMP_activity_antifungal = df_DRAMP_activity[df_DRAMP_activity['Activity'].str.contains("Antifungal|Antifunga|Antifugal", case=False, na=False)].copy()
df_DRAMP_activity_antifungal.head()
```

```python
df_DRAMP_activity_antifungal.shape
```

### Creamos una columna llamada label y asignamos 1, considerando que todas estas secuencias presentan la actividad antifungal

```python
df_DRAMP_activity_antifungal['label'] = 1
df_DRAMP_activity_antifungal
```

### Finalmente solo nos quedamos con las columnas sencuencia y label

```python
df_DRAMP_antifungal = df_DRAMP_activity_antifungal[["Sequence", "label"]]
df_DRAMP_antifungal.head()
```

#Limpieza de la Data

Verificación de duplicados

```python
df = df_DRAMP_antifungal

# 1. Crear el DataFrame que solo contiene las filas con secuencias duplicadas
filas_con_secuencias_duplicadas = df[
    df.duplicated(subset=['Sequence'], keep=False)
].sort_values(by='Sequence') # Ordenamos para que las secuencias repetidas aparezcan juntas

# 2. Imprimir los resultados
if len(filas_con_secuencias_duplicadas) > 0:
    print(f"✅ Se encontraron {len(filas_con_secuencias_duplicadas)} filas con secuencias duplicadas (incluyendo la primera aparición).")
    print("A continuación se muestran las filas duplicadas:")

    # Mostramos las columnas relevantes para la secuencia y actividad
    print(filas_con_secuencias_duplicadas[['Sequence', 'label']].head(20))
else:
    print("🎉 ¡No se encontraron secuencias duplicadas en tu DataFrame!")
```

```python
df_sin_duplicados = df_DRAMP_antifungal.drop_duplicates(subset=['Sequence'], keep='first')
df_sin_duplicados.shape
```

###Verificación de secuencias no canónicas

```python
df_sin_duplicados = df_sin_duplicados.copy()
df_sin_duplicados["Sequence"] = df_sin_duplicados["Sequence"].str.strip() #eliminamos los posibles espacios en la secuencia
```

```python
valid_mask = df_sin_duplicados["Sequence"].str.fullmatch(r"[ACDEFGHIKLMNPQRSTVWY]+", na=False)
invalid_df = df_sin_duplicados[~valid_mask].copy()
invalid_df
```

```python
# Mantener solo secuencias canónicas
filtered = df_sin_duplicados[valid_mask].copy()
filtered.shape
```

```python
df_DRAMP_canonicos= filtered
```

#Longitud de la secuencia

```python
df_DRAMP_canonicos = df_DRAMP_canonicos.copy()  # opcional pero limpio
df_DRAMP_canonicos["Largo"] = df_DRAMP_canonicos["Sequence"].str.len()
df_DRAMP_canonicos["Largo"].describe()
```

```python
sns.histplot(data=df_DRAMP_canonicos, x="Largo", fill=False, kde=True)
```

### Descargamos CSV final

```python
dataframe_a_descargar = df_DRAMP_canonicos
dataframe_a_descargar.to_csv(
    'DRAMP_antifungal.csv',  # Nombre que tendrá el archivo
    index=False                        # Importante: No incluir el índice de filas de Pandas
)
files.download('DRAMP_antifungal.csv')
```
